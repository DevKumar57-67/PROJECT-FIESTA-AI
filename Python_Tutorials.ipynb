```python
# Recursive Function is a function which calls itself recursively to solve specific problem
# It is generealy used in mathematical problems and divide and conquer problems

def fact(n):
  if n == 0:
    return 1
  else:
    return n * fact(n-1)

n = int(input("Enter the value of factorial you want to calculate:"))
print(fact(n))

```

```python
# LINKED LISTS IN PYTHON
# A linkedin list is a collection of elements called nodes where
# • each node stores data
# • each node stores a reference pointer to next node

# Exactly like a Train
# We have 3 types of Linked List
# - Singly Linked List
# - Doubly Linked List
# - Circular Linked List
# In Python we use class concept for LinkedIn lists
```

```python
# A python Linked List Structure

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
# An actual linked list

class Node:
    def __init__(self, new_data):
        self.data = new_data
        self.next = None


# Putting values in the LL

# First value
head = Node(10)

# Second value
head.next = Node(20)

# Third Value
head.next.next= Node(30)

# Fourth Value
head.next.next.next = Node(40)

# Printing the list

temp = head
while temp is not None:
    print(temp.data, end= " ")
    temp = temp.next
```
