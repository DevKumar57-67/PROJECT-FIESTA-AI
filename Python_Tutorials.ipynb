```python
# Python functions
# Python has two types of functions user defined functions and built in functions
# Python Functions are a block of statements that does a specific task. The idea is to put some commonly or repeatedly done task together and make a function so that instead of writing the same code again and again for different inputs, we can do the function calls to reuse code contained in it over and over again.
# Basic Python function to calculate sum of two numbers
# Python Functions uses def keyword with name of the function and parameters and return statement

def sum(a,b):
  return a+b


a = int(input("Enter the value of a:"))
b = int(input("Enter the value of b:"))

add = sum(a,b)
print("The sum is: ", add)
```

```python
# Function within Functions
# A function defined inside another function is called an inner function (or nested function). It can access variables from the enclosing function’s scope and is often used to keep logic protected and organized

# A basic nested function
def f1():
    s = 'I love GeeksforGeeks'
    def f2():
        print(s)

    f2()
f1()
```

```python
# Anomyous Function is a function without a name
# Format of the lambda function lambda arguments: expression

add = lambda a,b: a+b
a = int(input("Enter the value of a: "))
b = int(input("Enter the value of b: "))

print(add(a,b))
```

```python
# Recursive Function is a function which calls itself recursively to solve specific problem
# It is generealy used in mathematical problems and divide and conquer problems

def fact(n):
  if n == 0:
    return 1
  else:
    return n * fact(n-1)

n = int(input("Enter the value of factorial you want to calculate:"))
print(fact(n))

```

```python
# CLASSES
''' classes are considered as the collection of objects
A class defines a set of attributes and methods that the created objects (instances) can have. Some points on Python class:

Classes are created by keyword class.
Attributes are the variables that belong to a class.
Attributes are always public and can be accessed using the dot (.) operator. Example: Myclass.Myattribute'''


# A Python class

class dev:

  def __init__(self, name, age):
    self.name = name
    self.age = age


dev1= dev("Buddy", 9)
dev2 = dev("Buddy2", 10)

print(dev1.name)
print(dev1.age)
```
